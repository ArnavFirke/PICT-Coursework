```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/titanic.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin','Embarked'], axis=1)
```

```python
sns.countplot(x='Pclass', data=df,hue="Sex")
plot.xlabel('Passenger Class')
plot.title("Passenger Class Count")
```

```python
sns.histplot(data=df, x="Sex")
```

```python
sns.countplot(x='Sex',data=df,hue='Survived')
plot.show()
```

```python
plot.figure(figsize=(12, 8))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plot.title('Survival Count by Class')
plot.xlabel('Passenger Class')
plot.ylabel('Count')
plot.show()
```

```python
sns.histplot(df['Age'].dropna(), bins=30,kde=True)
plot.title('Distribution of Age')
plot.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
```

```python
sns.scatterplot(x='Age', y='Fare', data=df)
```

```python
sns.histplot(df['Fare'], bins=30, kde=True, color='skyblue')
plot.title('Distribution of Ticket Prices')
plot.xlabel('Fare')
plot.ylabel('Frequency')
plot.show()
```

```python
sns.set(style="whitegrid")

plot.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df)
plot.title('Boxplot of Age by Gender and Survival')
plot.show()
```
