```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings("ignore")
```

```python
iris=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/Iris.csv',encoding="ISO-8859-1") #reading csv file using pandas
print(iris.head())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
res=le.fit_transform(iris['Species'])

mapping = dict(zip(le.classes_, range(len(le.classes_))))
print("Mapping: ",mapping)

iris['Species']=res

```

```python
X=iris.drop(['Id','Species'],axis=1)
y=iris.Species
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
nb=GaussianNB()
mdl=nb.fit(X,y)
pred=mdl.predict(X)
print(accuracy_score(y,pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y, pred))
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y,pred)
print(matrix)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
conf_matrix = ConfusionMatrixDisplay(confusion_matrix=matrix)
conf_matrix.plot()
```
